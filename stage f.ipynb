```python
# modeling
X=df['Humidity'].values.reshape(-1,1)
y=df['Temperature (C)'].values.reshape(-1,1)
```

```python
 from sklearn.model_selection import train_test_split
```

```python
#stratify since respond vaviables are not balance
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
X_train.reshape(-1,1)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest_model = RandomForestRegressor(n_estimators = 500, random_state = 0)
```

```python
y_train.unique()
```

```python
X_train
```

```python
forest_model.fit(X_train,y_train)
```

```python
pred =forest_model.predict(X_test)
pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae = mean_absolute_error(y_test, pred)
round(mae, 3)
```

```python
plt.figure(figsize =(16,5))
plt.scatter(X_train,y_train , c = 'black')
plt.plot(pred,y_test, c= 'blue')
plt.show()
```
